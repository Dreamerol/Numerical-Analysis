## Задача 1

Да се намери $n$ така, че $n$-тата съставна
квадратурна формула на **правоъгълниците**/**трапеците**/**Симпсън**
да приближава интеграла
$\displaystyle
I = \int_{0}^{1} \frac{dx}{1+x}
$ с грешка, не по-голяма от $\displaystyle \varepsilon = 0.00001.$

Да се намери това приближение.


```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
#formula na rectangles
np.sqrt(10**5/12)
print(np.sqrt(10**5/12))
n = 92
def f(x):
    return 1/(1+x)
a = 0
b = 1
xmesh = np.linspace(a,b,n+1)
approximateIntegralRectangles = 0
for i in range(1,n+1):
    #from 0 to n 
    approximateIntegralRectangles += f((xmesh[i] + xmesh[i-1])/2)
approximateIntegralRectangles *= (b-a)/n
print(approximateIntegralRectangles)
x=sp.symbols('x')
print(sp.integrate(1/(1+x), (x,0,1)))
print(np.log(2))
```

### Задача 2

По метода на неопределените коефициенти да се намери формулата на Гаус-Льожандър с два възела.

```python
x,x1,x2,A1,A2 = sp.symbols('x x1 x2 A1 A2')

sp.solve(
    [
        sp.Eq(sp.integrate(1,(x,-1,1)), A1+A2),
        sp.Eq(sp.integrate(x, (x,-1,1)), A1*x1+A2*x2),
        sp.Eq(sp.integrate(x**2, (x,-1,1)), A1*x1**2+A2*x2**2),
        sp.Eq(sp.integrate(x**3, (x,-1,1)), A1*x1**3+A2*x2**3)
    ]
)
```

### Формули на Гаус-Льожандър:

| Брой възли | Коефициенти                 | Възли                          | АСТ |
|-----------|-----------------------------|--------------------------------|-----|
| 1         | A₁ = 2                      | x₁ = 0                         | 1   |
| 2         | A₁ = 1<br>A₂ = 1            | x₁ = −1/√3<br>x₂ = 1/√3        | 3   |
| 3         | A₁ = 5/9<br>A₂ = 8/9<br>A₃ = 5/9 | x₁ = −√(3/5)<br>x₂ = 0<br>x₃ = √(3/5) | 5   |
