```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.5*x + 0.2

x = np.linspace(-2, 6,100)
plt.scatter([0,1,2,3,4], [0,1,1,2,2])
plt.plot(x, f(x))
```

```python
import sympy as sp
a, b = sp.symbols('a,b')
def Phi(a,b):
    return (b)**2 + (a+b-1)**2 + (2*a+b-1)**2 + (3*a+b-2)**2 + (4*a+b - 2)**2 
sp.diff(Phi(a,b), a)
sp.diff(Phi(a,b),b)
sp.solve([sp.Eq(sp.diff(Phi(a,b), a),0), sp.Eq(sp.diff(Phi(a,b),b),0)])
```
