## Задача 1
Да се пресметне приближено стойността на интеграла
$\displaystyle I= \int_{0}^{4} \left( \frac{x^4}{8} - x^3 + 7.01x^2 - 16.05x + 12.06 \right)\, dx
$, където $f(x) =  \displaystyle \frac{x^4}{8} - x^3 + 7.01x^2 - 16.05x + 12.06 $, като се използват формулите на Гаус–Льожандър с 2 и 3 възела с подходяща смяна на променливите. Да се сравни резултатът с точната стойност. Да се сравни със стойностите, получени по формулите
на трапеците и на Симпсън.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

a=0
b=4

def f(x):
    return x**4/8 - x**3 +7.01*x**2 - 16.05*x + 12.06

def phi(t):
    return f((a+b)/2 + (b-a)/2*t)*(b-a)/2
x = sp.symbols('x')
sp.integrate(f(x), (x, 0, 4))
print("Gausse-Lagander 2 nodes: ", phi(-np.sqrt(3)/3) + phi(np.sqrt(3)/3))
print("Gausse-Lagander 3 nodes: ", phi(-np.sqrt(3/5))*5/9 + phi(0)*8/9 + phi(np.sqrt(3/5))*5/9)

print("Trapezoids: ", (b-a)/2*f((a)+f(b)))
print("Simpson: ", (b-a)/6 * (f(a) + f(b) + 4*f((a+b)/2)))
```

```python
a=0
b=2

eps = 10**-5
maxIterations = 50
iterations = 1
def f(x):
    return np.exp(x)-3

def derivative_f(x):
    return np.exp(x)
xcurrent = 2
xnext = (xcurrent - f(xcurrent)/derivative_f(xcurrent))
while(iterations <= maxIterations and abs(xnext-xcurrent) > eps):
    
    xcurrent = xnext 
    xnext = (xcurrent - f(xcurrent)/derivative_f(xcurrent))
    iterations+=1
print(iterations)
print("xnext: ", xnext, "xcurrent: ", xcurrent)
```
