```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1/(x+1)
```

```python
def l(x):
    return 1-x/2
def estimate_error(x):
    return abs(x*(x-1))
```

```python
x = np.linspace(0,1,100)
plt.plot(x, f(x), x, l(x))
plt.show()
plt.plot(x, abs(f(x) - l(x)))
plt.plot(x, estimate_error(x))
#plt.plot(x, estimate_error)
```

```python
def divided_difference(nodes, values, x):
    n = len(nodes)
    if n == 1:
        return values
    return (divided_difference(nodes[1:], values[1:], x) - divided_difference(nodes[0:n-1], values[0:n-1], x))/(nodes[n-1] - nodes[0])
```

```python
x = np.linspace(0, 5, 100)
nodes = np.array([0,1,4])
values = np.array([2, 5, 48])
print(divided_difference(nodes, values, x))
```

```python
def newton_Poly(nodes, values, x):
    p = 0.0
    n = len(nodes)
    for i in range(0, n):
        q = divided_difference(nodes[0:i+1], values[0:i+1], x)
        for j in range(0, i):
             q *= (x-nodes[j]) 
       # print(q)
        
        p += q

    return p[0]
        
```

```python
nodes = np.array([0, 1, 4])
values = np.array([2, 5, 48])

x = 20.0
print(newton_Poly(nodes, values, x))
```
