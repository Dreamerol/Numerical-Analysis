```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi, 100)
x0 = 0
x1 = np.pi / 6
x2 = np.pi/3
x3 = np.pi/2

y0 = np.sin(x0)
y1 = np.sin(x1)
y2 = np.sin(x2)
y3 = np.sin(x3)

```

```python
def giveX(i):
    if i == 0:
        return x0
    elif i == 1:
        return x1
    elif i == 2:
        return x2
    else:
        return x3
```

```python
def giveY(i):
    if i == 0:
        return y0
    elif i == 1:
        return y1
    elif i == 2:
        return y2
    else:
        return y3
```

```python
def y(idx):
    y = 1
    for i in range(0, 3):
        if idx == i:
            continue
        y *= ((x-giveX(i)) /(giveX(idx) - giveX(i)))
    return y
  
    
```

```python
#int x = np.linspace(-100, 100, 100)
def p(x):
    p = 0
    for i in range(0,3):
    
        p += y(i)* giveY(i)
    return p
      
```

```python
plt.plot(x, p(x))
```

```python

l0 = (x-x1)*(x-x2)*(x-x3) / ((x0 - x1)*(x0-x2)*(x0-x3))
l1 = (x-x0)*(x-x2)*(x-x3) / ((x1 - x0)*(x1-x2)*(x1-x3))
l2 = (x-x1)*(x-x0)*(x-x3) / ((x2 - x1)*(x2-x0)*(x2-x3))
l3 = (x-x1)*(x-x2)*(x-x0) / ((x3 - x1)*(x3-x0)*(x3-x2))
```

```python
plt.plot(x, l0)
plt.plot(x, l1)
plt.plot(x, l2)
plt.plot(x, l3)
```

```python

p = l0 * y0 + l1*y1 + l2*y2 + l3*y3
```

```python
plt.plot(x, p)
plt.plot(x, np.sin(x))
plt.scatter([x0,x1,x2,x3], [y0,y1,y2,y3])
```

```python
def basisPoly(idx, nodes, x):
    y = 1
    for i in range(0, len(nodes)):
        if i == idx:
            continue
        y *=((x - nodes[i])/(nodes[idx] - nodes[i]))
    return y
```

```python
def g(x):
    return np.exp(x)
#wwe must define the func with def!!!!!!!!!!!!!!!!!! or it wont be a np.array
```

```python
def Lagrange_poly(f, nodes, x):
    # nodesY = []
    # for i in range(0, len(nodes)):
    #     nodesY.append(f(nodes[i]))
    p = 0
    
    for i in range(0, len(nodes)):
        p += f(nodes[i])*basisPoly(i, nodes, x)
    return p
```
