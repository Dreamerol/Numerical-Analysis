```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.exp(x)
```

```python
def g(x):
    return 1+x+0.5*x**2 + 0.1667*x**3
```

```python
def absoluteError(x):
    return f(x) - g(x);
```

```python
def RelativeError(x):
   return absoluteError(x) / f(x)
```

```python
x = np.linspace(-0.5,0.5, 100)
plt.plot(x, f(x), x, g(x))
plt.show()
```

```python
plt.plot(x, absoluteError(x), x, RelativeError(x))
plt.legend(['f(x)', 'g(x)'])
plt.title("Absolute vs Relative error")
plt.show()
```

```python
def t(x):
    return (x-2)**7
```

```python
x = np.linspace(1.5, 2.5, 1000)
plt.plot(x, t(x))
plt.show()
```
