```python
#zadacha 1
time = np.array([0,1,1.5,2,2.5])
acc = np.array([0,1,1.5, 4,2])
T = 3
def basisPol(x):
    return np.array([1, np.cos(x), np.sin(x), np.cos(2*x), np.sin(2*x)])

A = np.zeros([5,5])
for i in range(5):
    A[i] = basisPol(time[i] * 2*np.pi/T)

sol = np.linalg.solve(A, acc)

def f(x):
    return sol[0]*1 + sol[1]*np.cos(x) + sol[2]* np.sin(x) + sol[3] * np.cos(2*x) + sol[4] * np.sin(2*x)
x = np.linspace(0,3,100)
plt.plot(x, f(x*2*np.pi/T))
plt.scatter(time, acc)
```

```python
#zadacha 1.1 4pi
time = np.array([0,1,1.5,2,2.5])
acc = np.array([0,1,1.5, 4,2])
T = 4*np.pi
def basisPol(x):
    return np.array([1, np.cos(x/2), np.sin(x/2), np.cos(x), np.sin(x)])

A = np.zeros([5,5])
for i in range(5):
    A[i] = basisPol(time[i])

sol = np.linalg.solve(A, acc)

def f(x):
    return sol[0]*1 + sol[1]*np.cos(x/2) + sol[2]* np.sin(x/2) + sol[3] * np.cos(x) + sol[4] * np.sin(x)
x = np.linspace(0,3,100)
plt.plot(x, f(x))
plt.scatter(time, acc)
```

```python
#zadacha 2
import sympy as sp
a, b = sp.symbols('a,b')
nodes = np.array([2,4,6,7,10,11,14,17,20])
values = np.array([4,5,6,7,8,8,11,10,12])
def f(nodes, values, a, b):
    sum = 0
    for i in range(len(values)):
        sum += (a*nodes[i] + b - values[i])**2
    return sum

eq1 = sp.diff(f(nodes,values,a,b), a)
eq2 = sp.diff(f(nodes,values,a,b), b)
sol = sp.solve([sp.Eq(eq1, 0), sp.Eq(eq2, 0)])
print(sol)

def g(x):
    return sol[a]*x + sol[b]

x = np.linspace(2, 20, 1000)
plt.plot(x, g(x))
```

```python
#zad 9
nodes = np.array([0,0.1, 0.2, 0.3 ,0.4])
values = np.array([1, 8, 4, 3.5, 5])
def basisPol(x,nodes, idx):
    pol = 1
    for i in range(len(nodes)):
        if idx == i:
            continue
        pol *= (x-nodes[i]) / (nodes[idx] - nodes[i])
    return pol

def lagrangePoly(x, nodes, values):
    poly = 0
    for i in range(len(nodes)):
        poly += values[i] * basisPol(x,nodes,i)
    return poly

n = 5
a = 0
b = 0.4
sum = 0
for i in range(1, n):
    sum += lagrangePoly(nodes[i-1],nodes,values) + 4 *lagrangePoly(((nodes[i-1] + nodes[i]) / 2), nodes,values) + lagrangePoly(nodes[i],nodes,values)
sum *= (nodes[i]-nodes[i-1])/(6)
print(sum)

a,b,x = sp.symbols('a,b,x')
print(sp.integrate(lagrangePoly(x,nodes,values), (x,0,0.4)))
```

```python
#zad 10
def f(x):
    return 1-1/np.tan(x)

def calcIntRect(e):
    n = math.ceil(np.sqrt(2*np.cos(1)/(np.sin(1)**3*12*e)))
    xmesh = np.linspace(-2,-1,n+1)
    sumi = 0
    a = -2
    b = -1
    for i in range(1,n+1):
        sumi += f((xmesh[i-1] + xmesh[i])/2)  
    sumi *= (b-a)/n
    return sumi


def calcIntTrap(e):
    n = math.ceil(np.sqrt(2*np.cos(1)/(np.sin(1)**3*12*e)))
    xmesh = np.linspace(-2,-1,n+1)
    sumi = 0
    a = -2
    b = -1
    for i in range(1,n+1):
        sumi += f(xmesh[i-1]) + f(xmesh[i]) 
    sumi *= (b-a)/(2*n)
    return sumi

def calcIntSimpson(e):
    n = math.ceil((2*np.cos(1)*(5+10*np.cos(1)**2)/(np.sin(1)**5*2880*e))**(1/4))
    xmesh = np.linspace(-2,-1,n+1)
    sumi = 0
    a = -2
    b = -1
    for i in range(1,n+1):
        sumi += f(xmesh[i-1]) + 4*f((xmesh[i] + xmesh[i-1])/2) + f(xmesh[i]) 
    sumi *= (b-a)/(6*n)
    return sumi
print(calcIntRect(0.001))
print(calcIntTrap(0.001))
print(calcIntSimpson(0.001))
x= sp.symbols('x')
fsym = 1 - 1/sp.tan(x)
print(sp.integrate(fsym, (x,-2,-1)))
```
