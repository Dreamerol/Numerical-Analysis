```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(x):
    return sp.exp(x)
a,b,c=sp.symbols('a,b,c')
x = sp.symbols('x')
def raz(x,a,b,c):
    return (f(x) - a*x**2 - b*x-c)**2
def phi(a,b,c):
    return sp.integrate(raz(x,a,b,c), (x,0,1.5))

eq1=sp.diff(phi(a,b,c), a)
eq2=sp.diff(phi(a,b,c), b)
eq3=sp.diff(phi(a,b,c), c)

dict = sp.solve([sp.Eq(eq1, 0), sp.Eq(eq2, 0), sp.Eq(eq3, 0)])
print(dict)
x = np.linspace(0,1.5,100)
def g(x):
    return dict[a]*x**2 + dict[b]*x+dict[c] 
    #convert to float
plt.plot(x, g(x))
plt.plot(x,np.exp(x))
```
